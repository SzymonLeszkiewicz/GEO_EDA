```python
import pandas as pd 
import os
```

```python
path = 'input_data'
```

```python
for file in sorted(os.listdir(path)):
    df = pd.read_csv(os.path.join(path, file))
    new_file = file[:-4]
    # save to parquet
    df.to_parquet(f'data/{new_file}.parquet', index=False)
    # print(df.shape, df.columns)
```
