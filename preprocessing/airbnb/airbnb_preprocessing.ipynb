```python
all= pd.read_csv(os.path.join(DATA_DIR,'listings_all.csv'))
all.head()
```

```python
all.isna().sum()
```

```python
null_percentage = all.isnull().sum()/all.shape[0]*100

# Below code gives list of columns having more than 60% null
col_to_drop = null_percentage[null_percentage>90].keys()

all_corrected = all.drop(col_to_drop, axis=1)
```

```python
all_corrected.isna().sum()
```

```python
all_corrected=all_corrected.drop(["neighbourhood_group", "license"], axis=1)
```

```python
all_corrected.info()
```

```python
all_corrected['name'] = all_corrected['name'].astype("string")
all_corrected['host_name'] = all_corrected['host_name'].astype("string")
all_corrected['neighbourhood'] = all_corrected['neighbourhood'].astype("string")
all_corrected['room_type'] = all_corrected['room_type'].astype("string")
all_corrected['last_review'] = all_corrected['last_review'].astype("string")
all_corrected["city"] = all_corrected["city"].astype("string")
all_corrected["date"] = all_corrected["date"].astype("string")
all_corrected["price"] = all_corrected["price"].astype("string")
```

```python
all_corrected.info()
```

```python
all_corrected["price"] = all_corrected["price"].str.replace(r'$', '')
all_corrected["price"] = all_corrected["price"].str.replace(r',', '')
```

```python
all_corrected["price"] = all_corrected["price"].astype("float")
```

```python
all_corrected.info()
```
