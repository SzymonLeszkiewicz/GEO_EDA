```python
import os 
import pandas as pd 
```

```python
current_working_directory = os.getcwd()
DATA_DIR=os.path.join(current_working_directory, 'data')
SUBSET_DIR='tokyo_newyork'
```

Tokyo&New York Checkins

```python
tokyo= pd.read_csv(os.path.join(DATA_DIR,SUBSET_DIR,'dataset_TSMC2014_TKY.csv'))
new_york= pd.read_csv(os.path.join(DATA_DIR,SUBSET_DIR,'dataset_TSMC2014_NYC.csv'))
```

```python
tokyo.head()
```

```python
from dateutil import parser
tokyo['date'] = tokyo['utcTimestamp'].apply(lambda x: parser.parse(x).date())
tokyo['time'] = tokyo['utcTimestamp'].apply(lambda x: parser.parse(x).time())
new_york['date'] = new_york['utcTimestamp'].apply(lambda x: parser.parse(x).date())
new_york['time'] = new_york['utcTimestamp'].apply(lambda x: parser.parse(x).time())

```

```python
print(len(tokyo))
print(len(new_york))
```

```python
new_york=new_york.dropna(subset=['latitude','longitude','venueCategory','date','time', 'userId'])
tokyo=tokyo.dropna(subset=['latitude','longitude','venueCategory','date','time','userId'])
```

```python
print(len(tokyo))
print(len(new_york))
```

```python
import numpy as np
columns = new_york.columns.values.tolist()
columns.remove('userId')
columns  
```

```python
frames=[]
for column in columns:
    grouped = new_york.groupby('userId')[column].apply(np.array).reset_index(name=column)
    frames.append(grouped)

```

```python
from functools import reduce
final_new_york = reduce(lambda df1,df2: pd.merge(df1,df2,on='userId'), frames)
final_new_york.head()
```

```python
frames=[]
for column in columns:
    grouped = tokyo.groupby('userId')[column].apply(np.array).reset_index(name=column)
    frames.append(grouped)

final_tokyo = reduce(lambda df1,df2: pd.merge(df1,df2,on='userId'), frames)
final_tokyo.head()
```

```python
final_new_york['city']='New York'
final_tokyo['city']='Tokyo'
final_tokyo.drop(['timezoneOffset'],axis=1,inplace=True)
final_new_york.drop(['timezoneOffset'],axis=1,inplace=True)
final_new_york['timezoneOffset']=-240
final_tokyo['timezoneOffset']=540
```

```python
data=pd.concat([final_tokyo, final_new_york], ignore_index=True)
```

```python
data.head()
```

```python
data.to_parquet(os.path.join(DATA_DIR,SUBSET_DIR,'tokyo_newyork.parquet'))
```
