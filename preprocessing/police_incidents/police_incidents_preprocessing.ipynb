```python
import os 
import pandas as pd 
```

```python
current_working_directory = os.getcwd()
DATA_DIR=os.path.join(current_working_directory, 'data')
```

```python
incidents=pd.read_csv(os.path.join(DATA_DIR, 'Police_Department_Incident_Reports.csv'))
```

```python
incidents.head()
```

```python
incidents.isna().sum()
```

```python
null_percentage = incidents.isnull().sum()/incidents.shape[0]*100
# Below code gives list of columns having more than 70% null
col_to_drop = null_percentage[null_percentage>75].keys()

incidents_filtered = incidents.drop(col_to_drop, axis=1)
```

```python
incidents_filtered.isna().sum()
```

```python
incidents_filtered.head()
```

```python
incidents_filtered.info()
```

```python
incidents_filtered["Point"].fillna("None",inplace=True)
```

```python
for column in incidents_filtered.select_dtypes(include=['object']).columns:
    incidents_filtered[column] = incidents_filtered[column].astype("string")
```

```python
incidents_filtered.info()
```

```python
incidents_filtered["x"]=incidents_filtered["Point"].apply(lambda p:  p.replace("(","").replace(")","").split(' ')[1] if p!="None" else -1.0)
incidents_filtered["y"]=incidents_filtered["Point"].apply(lambda p:  p.replace("(","").replace(")","").split(' ')[2] if p!="None" else -1.0)
```

```python
incidents_filtered["x"]=incidents_filtered["x"].astype("float64")
incidents_filtered["y"]=incidents_filtered["y"].astype("float64")
incidents_filtered.drop("Point",axis=1,inplace=True)
```

```python
incidents_filtered.info()
```

```python
incidents_filtered.to_parquet(os.path.join(DATA_DIR, 'police_incidents.parquet'))
```
