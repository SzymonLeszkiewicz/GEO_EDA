# Imports

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import LineString
from tqdm.contrib.concurrent import process_map
```

# Data Loading

```python
df_gowalla = pd.read_parquet("data/gowalla.parquet")
```

```python
df_gowalla.head()
```

```python
df_gowalla["geometry"] = process_map(
    LineString, df_gowalla["geometry"], chunksize=1000, max_workers=20
)
```

```python
df_gowalla.head()
```

```python
gdf_gowalla = gpd.GeoDataFrame(
    data=df_gowalla, geometry="geometry", crs="EPSG:4326"
)
```

```python
gdf_gowalla["geometry"].head().explore(
    tiles="CartoDB positron",  # use "CartoDB positron" tiles
)
```
