# Imports

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import LineString
from tqdm.contrib.concurrent import process_map
```

# Data Loading

```python
df_brightkite = pd.read_parquet("data/brightkite.parquet")
```

```python
df_brightkite.head()
```

```python
df_brightkite["geometry"] = process_map(
    LineString, df_brightkite["geometry"], chunksize=1000, max_workers=20
)
```

```python
df_brightkite.head()
```

```python
gdf_brightkite = gpd.GeoDataFrame(
    data=df_brightkite, geometry="geometry", crs="EPSG:4326"
)
```

```python
gdf_brightkite["geometry"].head().explore(
    tiles="CartoDB positron",  # use "CartoDB positron" tiles
)
```
