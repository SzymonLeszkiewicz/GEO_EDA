```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
path = 'data/archive/'
files = []
for f in os.listdir(path):
    if f.endswith('.csv'):
        files.append(os.path.join(path, f))
files
```

```python
df_checkins = pd.read_csv(files[0], encoding = 'windows-1251')
df_tags = pd.read_csv(files[1], encoding = 'windows-1251')
df_tips = pd.read_csv(files[2], encoding = 'windows-1251')
```

```python
df_checkins
```

```python
plt.hist(df_checkins['user_ID'].value_counts(), bins = 100)
plt.title('Checkins per user')
plt.xlabel('User checkins')
plt.ylabel('Frequency')
plt.show()
```

```python
# checkins per venue
plt.hist(df_checkins['venue_ID'].value_counts(), bins = 100)
plt.title('Checkins per venue')
plt.xlabel('Venue checkins')
plt.ylabel('Frequency')
plt.show()
```
